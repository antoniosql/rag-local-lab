# 02 · Embeddings con Ollama e indexación en Qdrant

Ahora tomamos los chunks del laboratorio anterior y los convertimos en vectores para insertarlos en Qdrant.

## Objetivo
- reutilizar el chunking anterior,
- generar embeddings con `OllamaEmbeddings`,
- crear o recrear una colección,
- insertar los puntos y validarlos.

```python
import sys
from pathlib import Path

from dotenv import load_dotenv

load_dotenv(Path("..") / ".env")
sys.path.insert(0, str((Path("..") / "app").resolve()))

from settings import Settings
from rag.chunking import chunk_documents, load_documents
from rag.pipeline import RAGPipeline

settings = Settings.from_env()
pipeline = RAGPipeline(settings)
documents = load_documents(Path("../docs"))
chunks = chunk_documents(documents, chunk_size=settings.chunk_size, overlap=settings.chunk_overlap)
len(chunks)
```

## Paso 1 · Probar embeddings

```python
sample_vector = pipeline.embeddings.embed_query(chunks[0]["text"])
len(sample_vector)
```

## Paso 2 · Ingerir de extremo a extremo

```python
result = pipeline.ingest_directory(Path("../docs"), force_recreate=True)
result
```

## Paso 3 · Validar colecciones

```python
pipeline.store.list_collections()
```

## Paso 4 · Recuperar una muestra directa desde Qdrant

```python
hits = pipeline.retrieve("¿Qué herramientas necesito para montar la mesa Oslo?", top_k=3)
hits
```
