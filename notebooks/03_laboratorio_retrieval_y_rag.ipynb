# 03 · Retrieval y RAG con LangChain

En este notebook unimos retrieval y generación usando el pipeline local del repositorio.

## Objetivo
- recuperar evidencia relevante,
- inspeccionar el contexto construido,
- ejecutar la cadena de respuesta,
- diferenciar claramente retrieval y generación.

```python
import sys
from pathlib import Path

from dotenv import load_dotenv

load_dotenv(Path("..") / ".env")
sys.path.insert(0, str((Path("..") / "app").resolve()))

from settings import Settings
from rag.pipeline import RAGPipeline
from rag.prompting import build_context

settings = Settings.from_env()
pipeline = RAGPipeline(settings)
settings
```

## Paso 1 · Retrieval puro

```python
question = "¿Puedo devolver una mesa ya montada?"
hits = pipeline.retrieve(question, top_k=3)
hits
```

## Paso 2 · Construir el contexto

```python
context = build_context(hits)
print(context[:2000])
```

## Paso 3 · Ejecutar la cadena LangChain

```python
result = pipeline.ask(question, top_k=3)
result
```

## Paso 4 · Comparar preguntas

```python
questions = [
    "¿Qué herramientas necesito para montar la mesa Oslo?",
    "¿Qué hago si faltan tornillos?",
    "¿Cuál es el plazo de entrega de la mesa Oslo?",
]

[pipeline.ask(q, top_k=3) for q in questions]
```
